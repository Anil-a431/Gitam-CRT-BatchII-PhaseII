```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>

struct Node{
	int Data;
	struct Node *Next;
}*S,*L,*C;

void Push(){
	struct Node *nn;
	nn = (struct Node *)malloc(sizeof(struct Node));
	printf("\n Enter a value : ");
	scanf("%d",&nn->Data);
	nn->Next = NULL;
	if(S==NULL){
		S = L = nn;
	}
	else{
		nn->Next = S;
		S = nn;
	}
	printf("\n Pushed to the Stack...");
	getch();
}
void Pop(){
	if(S==NULL){
		printf("\n Stack Underflow...");
	}
	else{//Remvoe the First Node of Linked List
		C = S;
		S = S->Next;
		C->Next = NULL;
		C = NULL;
		printf("\n Value is Poped from Stack...");
	}
	getch();
}
void Display(){
	if(S==NULL){
		printf("\n Stack Underflow...");
	}
	else{
		printf("\n Stack Items are : ");
		printf("\n ----------------- \n");
		for(C=S;C!=NULL;C=C->Next){
			printf("%d ",C->Data);
		}
	}
	getch();
}


int main(){
	int x;
	S = L = C =  NULL;
	while(1){
		system("cls");
		printf("\n ----------------- ");
		printf("\n STACk LINKED LIST ");
		printf("\n 1. Push");
		printf("\n 2. Pop");
		printf("\n 3. Display");
		printf("\n 4. Exit");
		printf("\n Enter Your Choice : ");
		scanf("%d",&x);
		switch(x){
			case 1: Push();
			break;
			case 2: Pop();
			break;
			case 3: Display();
			break;
			case 4: printf("\n Program is Terminating...");
			return 0;
			break;
		}
	}
	return 0;
}
```

```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
struct Node{
	int Data;
	struct Node *Next;
}*F,*R,*C;

void Enqueue(){
	struct Node *nn;
	nn = (struct Node *)malloc(sizeof(struct Node));
	printf("\n Enter the value : ");
	scanf("%d",&nn->Data);
	nn->Next = NULL;
	if(F==NULL){
		F = R = nn;
	}
	else{
		R->Next = nn;
		R = nn;
	}
	printf("\n Value is Inserted to Queue...");
	getch();
}
void Dequeue(){
	if(F == NULL){
		printf("\n Queue is Empty ... ");
	}
	else{
		C = F;
		F = F->Next;
		C->Next = NULL;
		C = NULL;
	}
	printf("\n Value is removed from Queue ... ");
	getch();
}
void Display(){
	if(F==NULL){
		printf("\n Single Linked List Empty");
	}
	else{
		printf("\n Single Linked List : ");
		printf("\n ------------------ \n");
		for(C=F;C!=NULL;C=C->Next){
			printf("%d ",C->Data);
		}
	}
	getch();
}

int main(){
	int x;
	F = R = C = NULL;
	while(1){
		system("cls");
		printf("\n -------------------- ");
		printf("\n Queue with LL        ");
		printf("\n 1. Enqueue ");
		printf("\n 2. Dequeue ");
		printf("\n 3. Display ");
		printf("\n 4. Exit ");
		printf("\n Enter the Choice : ");
		scanf("%d",&x);
		switch(x){
			case 1: Enqueue();
			break;
			case 2: Dequeue();
			break;
			case 3: Display();
			break;
			case 4: printf("\n Program is Terminating...");
			return 0;
			break;
		}
	}
	return 0;
}
```
