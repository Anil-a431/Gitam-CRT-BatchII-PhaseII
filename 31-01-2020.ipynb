```python
//Reading the Print the 2D array
#include<stdio.h>
#include<math.h>
int main(){
	int a[10][10],r,c,n,fwds=0,revs=0;
	scanf("%d",&n);
	for(r=0;r<n;r++){
		for(c=0;c<n;c++){
			scanf("%d",&a[r][c]);
		}
	}
	for(r=0;r<n;r++){
		for(c=0;c<n;c++){
			if(r==c){
				fwds += a[r][c];
			}
			if(r+c == n-1){
				revs += a[r][c];
			}
		}
	}
	printf("%d",abs(fwds-revs));
	return 0;
}
```

```python
//Reading the Print the 2D array
#include<stdio.h>
#include<math.h>
int main(){
	int a[10][10],r,c,n,s=0;
	scanf("%d",&n);
	for(r=0;r<n;r++){
		for(c=0;c<n;c++){
			scanf("%d",&a[r][c]);
		}
	}
	
	for(r=0;r<n;r++){
		for(c=0;c<n;c++){
			if(r!=0 && c!=0 && r!=(n-1) && c!=(n-1))
			{
				s = s + a[r][c];
			}
		}
	}
	printf("%d",s);
	return 0;
}
```

```python
//Reading the Print the 2D array
#include<stdio.h>
#include<math.h>
int main(){
	int r,c,n;
	scanf("%d",&n);
	for(r=0;r<n;r++){
		for(c=0;c<n;c++){
			if(r == 0 || c == 0 || r == n-1 || c == n-1 || (r==c) || (r+c == n-1))
			{
				printf("* ");
			}
			else{
				printf("  ");
			}
		}
		printf("\n");
	}
	
	return 0;
}
```

```python
#include<stdio.h>
int main()
{
	int r,c,a[10][10],i,j;
	scanf("%d%d",&r,&c);
	for(i=1;i<=r;i++){
		for(j=1;j<=c;j++){
			scanf("%d",&a[i][j]);
		}
	}
	for(i=0;i<=(r+1);i++){
		for(j=0;j<=(c+1);j++){
			if(i == 0 || j ==0 || i == (r+1) || j ==(c+1))
			{
				printf("* ");
			}
			else{
				printf("%d ",a[i][j]);
			}
		}
		printf("\n");
	}
	return 0;
}
```

```python
#include<stdio.h>
int main(){
	int r,c,a[10][10],i,j;
	scanf("%d%d",&r,&c);
	for(i=0;i<r;i++){
		for(j=0;j<c;j++){
			scanf("%d",&a[i][j]);
		}
	}
	for(i=-1;i<=r;i++){
		for(j=-1;j<=c;j++){
			if(i==-1 || j == -1 || i == r || j == c){
				printf("* ");
			}
			else{
				printf("%d ",a[i][j]);
			}
		}
		printf("\n");
	}
	return 0;
}
```

```python
14.
Input:-
2 3
1 2 3
4 5 6
Output:-
$ $ $ $ $ $ $
$ * * * * * $
$ * 1 2 3 * $
$ * 4 5 6 * $
$ * * * * * $
$ $ $ $ $ $ $
```

```python
#include<stdio.h>
int main(){
	int r,c,a[10][10],i,j;
	scanf("%d%d",&r,&c);
	for(i=0;i<r;i++){
		for(j=0;j<c;j++){
			scanf("%d",&a[i][j]);
		}
	}
	for(i=-2;i<=r+1;i++){
		for(j=-2;j<=c+1;j++){
			if(i == -2 || j == -2 || i == (r+1) || j == (c+1)){
				printf("$ ");
			}
			else if(i == -1 || j == -1 || i == r || j == c)
			{
				printf("* ");
			}
			else{
				printf("%d ",a[i][j]);
			}
		}
		printf("\n");
	}
	return 0;
}
```
