```python
Linear Search:-
===============
	** Unique List and Duplicate List
	** Sorted List and Unsorted List
	Worst Case Algorithm -- O(n)	
	Best Case Algorithm  -- O(1)
	
	It will start search from 0th index and moves
	fwd till n-1 index.
	Even after all n-1 index are finished if the
	target item is  not found -- not found
	If the target item is find then you can
	break the current searching.
```

```python
#include<stdio.h>
int main(){
	int i,targetItem = 8,flag=0;
	int a[7] = {19,12,35,2,8,22,43};
	for(i=0;i<7;i++){
		if(a[i]==targetItem){
			flag = 1;
			break;
		}
	}
	if(flag==1){
		printf("Target is found");
	}
	else{
		printf("Target is not found");
	}
	return 0;
}
```

```python
Bubble Sort:-
=============
	** This Algotith will compare the adj items
	if first item is greater than second item
	then shift/swap the items.
	
	** Complex -- O(n^2)
    
    - Find the first large or second large from the given input list then 
    you need to repeat for 1 or 2 times of outer loop and n-1 times of inner
    loop
```

```python
#include<stdio.h>
void bubbleSort(int a[],int n){
	int i,j,temp;
	for(i=0;i<n-1;i++){
		for(j=0;j<n-1;j++){
			if(a[j]>a[j+1]){
				temp = a[j];
				a[j] = a[j+1];
				a[j+1] = temp;
			}
		}
	}
	for(i=0;i<n;i++){
		printf("%d ",a[i]);
	}
}
int main()
{
	int x[6] = {19,1,25,12,2,18};
	bubbleSort(x,6);
	return 0;
}
```

```python
Binary Search:-
---------------
** Unique List 
** Sorted List

-- Best case Algorithm -- O(1)
-- Worst case Algorithm -- Log(n)
-- Binary Search algorithm starts the search
from middle index and goes to left or right
direction based on the target value.
```

```python
#include<stdio.h>
int binarySearch(int a[],int n,int tar){
	int mid,leftIndex,rightIndex;
	leftIndex = 0;
	rightIndex = n-1;
	while(leftIndex<=rightIndex){
		mid = leftIndex + (rightIndex-leftIndex)/2;
		if(a[mid]==tar){
			return mid;
		}
		if(a[mid]>tar){
			rightIndex = mid - 1;
		}
		else{
			leftIndex = mid + 1;
		}
	}
	return -1;//Not found case
}
int main(){
	int res,x[8] = {1,12,19,24,56,87,88,95};
	res = binarySearch(x,8,121);
	if(res>=0){
		printf("Found");
		printf("\n Position = %d",res+1);
	}
	else{
		printf("\n Not found");
	}
	return 0;
}
```

```python
#include<stdio.h>
void sort(int a[],int n){
	int i,j,temp;
	for(i=0;i<n-1;i++){
		for(j=0;j<n-1-i;j++){
			if(a[j]>a[j+1]){
				temp = a[j];
				a[j] = a[j+1];
				a[j+1] = temp;
			}
		}
	}
}
int main(){
	int n,a[100],i,mid;
	scanf("%d",&n);
	for(i=0;i<n;i++){
		scanf("%d",&a[i]);
	}
	sort(a,n);
	mid = 0 + (n-0)/2;
	for(i=0;i<mid;i++){
		printf("%d ",a[i]);
	}
	for(i=n-1;i>=mid;i--){
		printf("%d ",a[i]);
	}
	return 0;
}
```

```python
Selection Sort:-
================

This Algorithm sorts a list(an array) by
repeatedly finding minimum element(item) and
swapping to first indexes.

Complex : O(n^2)
    
    - Find least item or second least item from the given list then
    use Selection Sort Algorithm with 1*n or 2*n respectively.
```

```python
#include<stdio.h>
void selectionSort(int a[],int n){
	int i,j,k,temp;
	for(i=0;i<3;i++){
		k = i;
		for(j=i+1;j<n;j++){
			if(a[j]<a[k]){
				k = j;//Holding least value Index
			}
		}
		//swap the indexes of k and i
		temp = a[i];
		a[i] = a[k];
		a[k] = temp;
	}
	for(i=0;i<n;i++){
		printf("%d ",a[i]);
	}
}
int main(){
	int x[6] = {19,3,25,1,16,22};
	selectionSort(x,6);
	return 0;
}
```

```python
STACK (Array):-
===============
** Memory Storage - Static or Compile time
** STACK -- Variables
** STACK -- Looks like single sided open box
** STACK -- LIFO(Last IN First OUT)
** Insert an Item to the Stack called PUSH
** Remove an item from the Stack called POP
** Represents the Top item of the stack called 
   PEEK
   
** STACK Implements with static -- Array
	dyanmic -- Linked List
```

```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
int main(){
	int a[5],top=-1,x,i;
	while(1){
		system("cls");
		printf("---------------------");
		printf("\n---- STACK --------");
		printf("\n---- 1. Push ------");
		printf("\n---- 2. Pop  ------");
		printf("\n---- 3. Peek ------");
		printf("\n---- 4. Display ---");
		printf("\n---- 5. Exit  -----");
		printf("\n Enter Your choice");
		scanf("%d",&x);
		switch(x){
			case 1: // Insert an Item to the Stack
					if(top<4){
						top++;
						printf("\n Enter a Value : ");
						scanf("%d",&a[top]);
						printf("\n Value is Pushed...");
						getch();
					}
					else{
						printf("\n Stack is Full \n Stack is Overflow");
						getch();
					}
			break;
			case 2: //Remove the Stack Item
					if(top!=-1){
						a[top] = 0;//Avoid the Garbage values
						top--;
						printf("\n Value is removed...");
						getch();
					}
					else{
						printf("\n Stack Empty \n Stack Underflow");
						getch();
					}
			break;
			case 3: // Top item of the stack
					if(top!=-1){
						printf("\n Peek Item of the Stack : ");
						printf("%d ",a[top]);
						getch();
					}
					else{
						printf("\n Stack Empty \n Stack Underflow");
						getch();
					}
			break;
			case 4: //Display the Stack Values
					if(top!=-1){
						printf("\n Stack Values are : ");
						printf("\n ------------------ \n");
						for(i=top;i>=0;i--){
							printf("%d ",a[i]);
						}
						getch();
					}
					else{
						printf("\n Stack is Empty \n Stack is Underflow");
						getch();
					}
			break;
			case 5: printf("\n The Program is termininating...");
			return 0;
			break;
		}
	}
	return 0;
}
```

```python
QUEUE (Array) :-
================

** Queue will two sided open box. 
** Values will be enters from Rear it means one direction
** Values will be removed from Front it means another direction
** Insert an item Queue is called as Enqueue
** Remove an item Queue is called as Dequeue
** By default the F = -1 and R = -1
** If any time F and R is Pointing to 0 means there is single item is present 
in the queue, else it have more than single item
```

```python
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
int main(){
	int a[5],F=-1,R=-1,x,i;
	while(1){
		system("cls");
		printf("\n --------------------- ");
		printf("\n ---- QUEUE ---------- ");
		printf("\n ---- 1. Insert ------ ");
		printf("\n ---- 2. Remove ------ ");
		printf("\n ---- 3. Display ----- ");
		printf("\n ---- 4. Exit  ------- ");
		printf("\n Enter the Choice : ");
		scanf("%d",&x);
		switch(x){
			case 1: //Insert an Item -- Enqueue
					if(R<4){
						R++;
						if(R==0){
							F = 0;
						}
						printf("\n Enter a Value : ");
						scanf("%d",&a[R]);
						printf("\n Value is Inserted...");
						getch();
					}
					else{
						printf("\n Queue is Full");
						getch();
					}
			break;
			case 2: 
					if(F != -1 && R != -1)
					{
						if(F==R){
							a[F] = 0;//Avoid the Garbage
							F = -1;
							R = -1;
							printf("\n Value is removed...");
							getch();
						}
						else{
							for(i=F+1;i<=R;i++){
								if(i==R){
									a[i-1] = a[i];//
									a[R] = 0;
									R--;
								}
								else{
									a[i-1] = a[i];
								}
							}
							printf("\n Value is removed...");
							getch();
						}
					}
					else{
						printf("\n Queue is Empty");
						getch();
					}
			break;
			case 3: //Display the Queue Items
					if(F != -1 && R != -1){
						printf("\n Queue Values are : ");
						printf("\n -------------- --- \n");
						for(i=F;i<=R;i++){
							printf("%d ",a[i]);
						}
						getch();
					}
					else{
						printf("\n Queue is Empty");
						getch();
					}
			break;
			case 4: printf("\n Program is Terminating...");
			return 0;
			break;
		}
	}
	return 0;
}
```
